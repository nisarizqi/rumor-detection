```python
import pandas as pd

# Baca dua file CSV
df1 = pd.read_csv('labelling_result.csv')
df2 = pd.read_csv('cleaned_distinct_data_final.csv')

# Gabungkan dua dataframe berdasarkan kolom tertentu
merged_df = pd.merge(df1, df2, on='tweet_url', how='inner') # Ubah 'inner' sesuai kebutuhan

# Simpan dataframe hasil gabungan ke dalam file CSV baru
merged_df.to_csv('complete_labelled_data.csv', index=False)
```
