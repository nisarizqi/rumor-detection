```python
# Concat data

import pandas as pd
import glob

# Specify the directory where your CSV files are located
directory = "tweets-data/"

# Get a list of all CSV files in the specified directory
csv_files = glob.glob(directory + "*.csv")

# Initialize an empty list to store DataFrames
dfs = []

# Read each CSV file into a DataFrame and append to the list
for csv_file in csv_files:
    df = pd.read_csv(csv_file, delimiter=";")
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs)

# Remove duplicates based on all columns
unique_df = combined_df.drop_duplicates()

# Display the unique data
print(unique_df)

# Save the unique data to a new CSV file
unique_df.to_csv('distinct_data.csv', index=False)
```

```python
import pandas as pd
import re

# Read the CSV file into a DataFrame
df = pd.read_csv('distinct_data.csv')

# Function to check if a URL is not a retweet or quote
def remove_mention_web_url(text):
    urls = re.findall(r'http\S+', text)  # Extract URLs from text
    for url in urls:
        if url.startswith("https://t.co"):
            return False
    return True

# Filter out tweets with URLs that are not retweets or quotes
filtered_df = df[~df['full_text'].apply(remove_mention_web_url)]

# Save the filtered DataFrame to a new CSV file
filtered_df.to_csv('filtered_url_data.csv', index=False)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Create stemmer for Bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Get stopwords for Bahasa Indonesia
stop_words = set(stopwords.words('indonesian'))

# Read the CSV file into a DataFrame
df = pd.read_csv('distinct-data.csv')

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove account mentions
    text = re.sub(r'@\w+', '', text)
    # Remove special characters and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenization
    tokens = nltk.word_tokenize(text)
    # Remove stopwords and stemming
    clean_tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    return ' '.join(clean_tokens)

# Apply the preprocess_text function to the desired column
df['preprocessed_column'] = df['full_text'].apply(preprocess_text)

# Display the DataFrame with the preprocessed column
print(df)

# Save the DataFrame back to a CSV file
df.to_csv('preprocessed_data.csv', index=False)
```
